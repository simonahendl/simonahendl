```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

x = np.linspace(0, 10, 100)
y1, y2, y3 = np.cos(x), np.cos(x + 1), np.cos(x + 2)
names = ['Signal 1', 'Signal 2', 'Signal 3']
```

```python
plt.figure(1)
plt.plot(x, y1)
plt.title(names[0])

plt.figure(2)
plt.plot(x, y2)
plt.title(names[1])

plt.figure(3)
plt.plot(x, y3)
plt.title(names[2])
```
